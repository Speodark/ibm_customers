```python
import pandas as pd
import vaex
```

```python
df = pd.read_csv('telco-customer-churn-by-IBM.csv')[
    [
        'customerID', 'gender', 'tenure', 'InternetService', 'Contract', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges',
        'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 
        'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn'
    ]
]
```

```python
df["Partner"] = df["Partner"].apply(lambda x: False if 'no' in x.lower() else True)
df["Dependents"] = df["Dependents"].apply(lambda x: False if 'no' in x.lower() else True)
df["PhoneService"] = df["PhoneService"].apply(lambda x: False if 'no' in x.lower() else True)
df["MultipleLines"] = df["MultipleLines"].apply(lambda x: False if 'no' in x.lower() else True)
df["OnlineSecurity"] = df["OnlineSecurity"].apply(lambda x: False if 'no' in x.lower() else True)
df["OnlineBackup"] = df["OnlineBackup"].apply(lambda x: False if 'no' in x.lower() else True)
df["DeviceProtection"] = df["DeviceProtection"].apply(lambda x: False if 'no' in x.lower() else True)
df["TechSupport"] = df["TechSupport"].apply(lambda x: False if 'no' in x.lower() else True)
df["StreamingTV"] = df["StreamingTV"].apply(lambda x: False if 'no' in x.lower() else True)
df["StreamingMovies"] = df["StreamingMovies"].apply(lambda x: False if 'no' in x.lower() else True)
df["PaperlessBilling"] = df["PaperlessBilling"].apply(lambda x: False if 'no' in x.lower() else True)
df["Churn"] = df["Churn"].apply(lambda x: False if 'no' in x.lower() else True)
```

```python
df["gender"] = df["gender"].astype("category")
df["SeniorCitizen"] = df["SeniorCitizen"].astype(bool)
df["Partner"] = df["Partner"].astype(bool)
df["Dependents"] = df["Dependents"].astype(bool)
df["PhoneService"] = df["PhoneService"].astype(bool)
df["MultipleLines"] = df["MultipleLines"].astype(bool)
df["OnlineSecurity"] = df["OnlineSecurity"].astype(bool)
df["OnlineBackup"] = df["OnlineBackup"].astype(bool)
df["DeviceProtection"] = df["DeviceProtection"].astype(bool)
df["TechSupport"] = df["TechSupport"].astype(bool)
df["StreamingTV"] = df["StreamingTV"].astype(bool)
df["StreamingMovies"] = df["StreamingMovies"].astype(bool)
df["PaperlessBilling"] = df["PaperlessBilling"].astype(bool)
df["Churn"] = df["Churn"].astype(bool)
df["TotalCharges"] = pd. to_numeric(df["TotalCharges"].str.strip())
```
